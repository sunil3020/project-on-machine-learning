```python
df_target=df_house_price['price']
df_feature= df_house_price.drop('price',axis=1)
```

```python
df_num = df_house_price.select_dtypes(include='number')
df_cat = df_house_price.select_dtypes(include='object')
```

```python
# use dummy encoding
df_encoded = pd.get_dummies(df_cat)
# reset the index
df_encoded = df_encoded.reset_index()
```

```python
# show the encoded data
df_encoded.head()
```

```python
# import the standard scalar
from sklearn.preprocessing import StandardScaler
```

```python
# intiate the standard scalar
ss=StandardScaler()
```

```python
# scale the value present in numerical data frame
scale = ss.fit_transform(df_num)
```

```python
# numeric the data into the dataframe
scaled_df=pd.DataFrame(scale,columns=df_num.columns)

# reset the index
scaled_df=scaled_df.reset_index()
```

```python
scaled_df.head()
```

```python
# concade encoded variable and scaled variable 
df_independent=pd.concat([df_encoded,scaled_df],axis=1)
```

```python
df_independent.head()
```

```python
# import the train test split
from sklearn.model_selection import train_test_split
```

```python
# split the data 
x_train,x_test,y_train,y_test= train_test_split(df_independent,df_target,test_size=0.2)
```

```python
# import the model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# initiate the model
LR=LinearRegression()
```

```python
# train the linear regresstion model
model_lr=LR.fit(x_train,y_train)
```

```python
# perform the prediction
predicted=model_lr.predict(x_test)
```

```python
# show the predected value
predicted
```
